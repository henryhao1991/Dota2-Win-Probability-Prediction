```python
#Custom python files
from dataloader import *
from model import heuristic
from training import *

#Other libraries
import numpy as np
from matplotlib import pyplot as plt
```

```python
heuristic = TrainingAndEvaluation(model='heuristic', collate_fn=PadSequence)
```

# Accuracy over 1500+ games at the end of the game

```python
threshold = 0.1
percentage = 0.05
acc = heuristic.get_accuracy(threshold=threshold, percentage=percentage)
```

```python
per = np.arange(0,1.0,percentage)
plt.figure(0)
plt.plot(per, acc)
plt.xlim(-0.05,1.05)
plt.xlabel('Percentage of the Game')
plt.ylabel('Accuracy')
plt.show()
```

# A sample game that heuristic model works decent

```python
sample_graph = heuristic.get_single_game_prediction(10)
```

```python
time_stamp = 0.5*np.arange(len(sample_graph))

plt.figure(1)
plt.plot(time_stamp, sample_graph.numpy())
plt.axhline(0.5, color='k', linestyle='-')
plt.ylim(-0.1,1.1)
plt.xlabel('time (minute)')
plt.ylabel('Radiant Win Probability')
plt.show()
```
